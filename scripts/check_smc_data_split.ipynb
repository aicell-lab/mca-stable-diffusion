# Check filter

The 4 criteria filter 4 uses: 1. "status" == 35 2. "Ab state" == 'IF_FINISHED' 3. str("gene_names") != "nan" 4. "sequences" is non-empty

```python
filter_func = lambda x: int( x['status']) == 35 and x['Ab state'] == 'IF_FINISHED' and str(x['locations']) != "nan"
filter_func2 = lambda x: int( x['status']) == 35 and x['Ab state'] == 'IF_FINISHED'
filter_func3 = lambda x: int( x['status']) == 35 and x['Ab state'] == 'IF_FINISHED' and str(x["gene_names"]) != "nan"
filter_func4 = lambda x: int( x['status']) == 35 and x['Ab state'] == 'IF_FINISHED' and str(x["gene_names"]) != "nan" and x["sequences"]
flt_data_info = list(filter(filter_func, data_info))
flt2_data_info = list(filter(filter_func2, data_info))
flt3_data_info = list(filter(filter_func3, data_info))
flt4_data_info = list(filter(filter_func4, data_info))
len(flt_data_info), len(flt2_data_info), len(flt3_data_info), len(flt4_data_info)
```

```python
flt2_idcs, flt3_idcs, flt4_idcs = [], [], []
for i, x in enumerate(data_info):
    if filter_func2(x):
        flt2_idcs.append(i)
    if filter_func3(x):
        flt3_idcs.append(i)
    if filter_func4(x):
        flt4_idcs.append(i)
len(flt2_idcs), len(flt3_idcs), len(flt4_idcs)
```

```python
for x in flt4_data_info:
    prot = x["gene_names"]
    if "," in prot:
        print(prot)
```

Check whether the indices in the data split are exactly the same as the filtered indices above.

```python
with open("/data/wei/hpa-webdataset-all-composite/stage1_data_split.json") as in_file:
    data_split = json.load(in_file)
data_split.keys()
```

```python
flt2_idcs_set = set(flt2_idcs)
flt3_idcs_set = set(flt3_idcs)
flt4_idcs_set = set(flt4_idcs)
train_idcs = data_split["train"]
valid_idcs = data_split["validation"]
print(len(set(train_idcs + valid_idcs)))
for i in (train_idcs + valid_idcs):
    if i not in flt3_idcs_set:
        print(i)
```

The previous data split used filter 3 instead of filter 4. I will refilter the indices.

```python
train_idcs_flt4 = set(train_idcs) & flt4_idcs_set
valid_idcs_flt4 = set(valid_idcs) & flt4_idcs_set
len(train_idcs_flt4), len(valid_idcs_flt4)
```

# Check split 
Check whether all the validation and test (antibody, cell line) pairs are in training as well

```python
train_prot_cl_pairs, valid_prot_cl_pairs = set(), set()
for i in train_idcs_flt4:
    x = data_info[i]
    train_prot_cl_pairs.add((x["gene_names"], x["atlas_name"]))
for i in valid_idcs_flt4:
    x = data_info[i]
    valid_prot_cl_pairs.add((x["gene_names"], x["atlas_name"]))
len(train_prot_cl_pairs), len(valid_prot_cl_pairs)
```

```python
assert valid_prot_cl_pairs.issubset(train_prot_cl_pairs)
```

The split looks good.

```python
data_split_flt4 = {"train": list(train_idcs_flt4), "validation": list(valid_idcs_flt4)}
with open("/data/wei/hpa-webdataset-all-composite/stage1_data_split_flt4.json", "w") as out_file:
    json.dump(data_split_flt4, out_file)
```
